## Data Cleaning

Before we begin analyzing the data and creating visualizations, the data could use some cleaning. The column names need cleaning and standardizing and the data will be easier to work with if we convert the columns to relevant types. First, since we're looking at only euro's exchange rate with the U.S. Dollar, first we'll split off the `Period\\Unit:` and `[US dollar ]` columns into a new dataframe, then rename them to `Time` and `US_Dollar`.  

*Note:  The project instructions did this in a different order. I rearranged it to split out the desired columns first because it made more sense to me.* 

```python
euro_to_dollar = exchange_rates[['Period\\Unit:','[US dollar ]']].copy()
euro_to_dollar.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
euro_to_dollar.head()
```

Now we'll investigate our values to make sure there are no surprises.

```python
euro_to_dollar['Time'].value_counts()
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

We know from running DataFrame.info() that there NA values in either column and both of their data types are "object". The `Time` values all look good, but they would be easier to work with as datetimes. In the`US_dollar` column, all of the values are numbers, except for the 62 containing '-'.   We can drop those rows and convert the rest to and float types. 

```python
    #Convert Time to datetime and sort ascending
euro_to_dollar['Time'] = pd.to_datetime(euro_to_dollar['Time'])
euro_to_dollar.sort_values('Time', inplace=True)
euro_to_dollar.reset_index(drop=True, inplace=True)

    # Remove rows in US_Dollar containing '-' and convert to float
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)

    #Check results
euro_to_dollar.info()  
```

```python
euro_to_dollar
```

## Visualizing the Data with a Rolling Mean

Now that our data is cleaned and ready, we'll create a line plot to visually inspect the change of the euro-dollar exchange rate over the full length of time in our data.

```python
style.use('fivethirtyeight')
# sns.set_style('white')
plt.figure(figsize = (8,4))


sns.lineplot(data = euro_to_dollar, x = 'Time',y = 'US_dollar')

plt.title('Euro to U.S. Dollar Exchange Rate 1999-2001')
plt.ylabel('Exchange Rate')
plt.xlabel('')

plt.show()
```

The graph is rather jagged due to the daily variations of the exchange rate.  We could pick values at longer intervals than one day, but that might mask interesting trends.  Instead, we'll use a rolling mean with a 30 day window.

```python
### Add Rolling_mean column
euro_to_dollar['Rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()

# Use fivethirtyeight style throughout workbook
plt.style.use("fivethirtyeight")

### Plot original data
fig, axes = plt.subplots(1,2,figsize = (10,4),tight_layout = True)
sns.lineplot(
    data = euro_to_dollar, 
    x = 'Time',
    y = 'US_dollar',
    ax = axes[0],
    lw = 1  # default is very heavy
).set(
    title = "Original",
    xlabel = '',
    ylabel = 'Exchange Rate'
)


### Plot rolling mean data
sns.lineplot(
    data = euro_to_dollar,
    x = 'Time',
    y = 'Rolling_mean',
    ax = axes[1],
    lw = 1 # default is very heavy
).set(
    title = "30-Day Rolling Mean",
    xlabel = '',
    ylabel = 'Exchange Rate'
)


plt.show()
```

Now we have a somewhat smoother graph and it's easier to see the overall trends without as much noise. It's time to decide on a story and how to present the data for our audience.  

## Comparing Exchange Rates During U.S. Presidencies

Our data spans the terms of the last three U.S. Presidents.  Let's take a look at how the exchange rate varied over those three terms. 

```python
### Isolate the years of each presidency 
bush =  euro_to_dollar[(euro_to_dollar['Time'] > '2001-01-20') & (euro_to_dollar['Time'] < '2009-01-20')]
obama = euro_to_dollar[(euro_to_dollar['Time'] > '2009-01-20') & (euro_to_dollar['Time'] < '2017-01-20')]
trump = euro_to_dollar[(euro_to_dollar['Time'] > '2017-01-20') & (euro_to_dollar['Time'] < '2021-01-20')]

#### create lists to loop through
presidencies = [bush,obama,trump]
labels = ["BUSH","OBAMA"," TRUMP"]
colors = ["red","blue","orange"]

### Set up the plot
style.use("fivethirtyeight")
plt.figure(figsize = (12,4))

# ### Create the charts
plt.title("EURO-USD rate increased under presidents G.W. Bush & Trump, decreased under Obama",
          fontsize=12, weight='bold')
for i in range(3):

        sns.lineplot(
            data = presidencies[i],
            x = 'Time',
            y = 'Rolling_mean',
            color = colors[i],
            lw = 1,
            label = labels[i]
        ).set(
        xlabel = '',
        ylabel = '',
        ylim = (.8,1.6),
        
        )

plt.legend()

plt.show()
```

We see that the Euro overall gained strength during Bush's presidency by a net of 0.76.  It then lost about 60% of that (0.48) during Obama's term, and made a modest recovery of 0.2 points during Trump's four years in office.  It is interesting to note that the euro generally got stronger during the two Republicans' terms while losing strength during the one Democrat's.  Without more data it's difficult to say whether this is a coincidence or if the party in power contributes to the relative strength of the euro and the dollar.  According to these articles on  [fxempire](https://www.fxempire.com/news/article/the-effect-of-us-presidents-on-eurusd-movement-393715) and [Walton College](https://walton.uark.edu/insights/the-almighty-dollar.php), these trends have existed since the advent of the euro, largely due to the tighter monetary policies of Republican administrations and the looser policies of the Democrats.
