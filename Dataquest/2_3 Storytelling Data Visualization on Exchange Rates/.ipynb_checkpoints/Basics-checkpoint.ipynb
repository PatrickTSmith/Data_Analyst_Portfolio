# Storytelling Data Visualization on Exchange Rates

## Introduction
Visualization and storytelling are an important part of data analysis.  In this notebook, we'll explore the daily Euro exchange rates from  1999 to 2021, specifically in relation to the U.S. Dollar, and decide the best way to visually tell a story about the data.

The [dataset](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020) we will use contains the daily exchange rate between the euro and various other world currencies.  Daria Chemkaeva compiled it from data made available by the European Central Bank. The dataset was downloaded in January 2021.  

Let's get our first look at the data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style

%matplotlib inline
```

```python
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
exchange_rates.info()
```

```python
exchange_rates.head()
```

```python
exchange_rates.tail()
```

The dataset contains 5699 rows and 41 columns.  Aside from the `Period\Unit:` column (which contains the date), the columns are all different currency exchange rates.  3 of them are float64 and the rest are object types. Most columns have no null values, but a little under half (17) do contain some null values.

## Data Cleaning

Before we begin analyzing the data and creating visualizations, the data could use some cleaning. The column names need cleaning and standardizing and the data will be easier to work with if we convert the columns to relevant types. First, since we're looking at only euro's exchange rate with the U.S. Dollar, first we'll separate split off the `Period\\Unit:` and `[US dollar ]` columns into a new dataframe, then rename them to `Time` and `US_Dollar`.  

*Note:  The project instructions did this in a different order. I rearranged it to split out the desired columns first because it made more sense to me.* 

```python
euro_to_dollar = exchange_rates[['Period\\Unit:','[US dollar ]']].copy()
euro_to_dollar.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
euro_to_dollar.head()
```

Now we'll investigate our values to make sure there are no surprises.

```python
print(euro_to_dollar['Time'].value_counts(),'\n\n',euro_to_dollar['US_dollar'].value_counts())
```

Our initial exploration showed that we have no NA values in either column and both of their data types are "object". The `Time` values all look good, but they would be easier to work with as datetimes. In the`US_dollar` column, all of the values are numbers, except for the 62 containing '-'.   We can drop those rows and convert the rest to and float types. 

```python
    #Convert Time to datetime and sort ascending
euro_to_dollar['Time'] = pd.to_datetime(euro_to_dollar['Time'])
euro_to_dollar.sort_values('Time', inplace=True)
euro_to_dollar.reset_index(drop=True, inplace=True)

    # Remove rows in US_Dollar containing '-' and convert to float
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)

    #Check results
euro_to_dollar.info()  
```

## Visualizing the Data with a Rolling Mean

Now that our data is cleaned and ready, we'll create a line plot to visually inspect the change of the euro-dollar exchange rate over the full length of time in our data.

```python
plt.plot(euro_to_dollar['Time'],euro_to_dollar['US_dollar'])
plt.title('Euro to U.S. Dollar 1999-2001')

plt.xlabel('Date')
plt.ylabel('Dollars Per Euro')

plt.show()

```

The graph is rather jagged due to the daily variations of the exchange rate.  We could pick values at longer intervals than one day, but that might mask interesting trends.  Instead, we'll use a rolling mean with a 30 day window. We can also loop over the data and plot each year separately in a different color to make it easier to read.

```python
euro_to_dollar['Rolling_mean'] = euro_to_dollar['US_dollar'].rolling(2).mean()



fig, ax = plt.subplots(2,1,figsize = (20,10),tight_layout = True)
ax[0].plot(euro_to_dollar['Time'],euro_to_dollar['US_dollar'])
ax[0].set_title('Original')

for year in range(1999,2021):
    year_data = euro_to_dollar[euro_to_dollar['Time'].dt.year == year]
    ax[1].plot(year_data['Time'],year_data['Rolling_mean'],label = year)
    ax[1].set_title('30-Day Rolling Mean')

plt.show()


```

Now we have a somewhat smoother graph and it's easier to see the overall trends without as much noise. It's time to decide on a story and how to present the data for our audience.  

## Coming Up With an Idea

We need to decide how to visually represent our data to our audience, which means telling a story.  Some ways we could do this include:
- Comparing the exchange rate by year.
- Showing how the rate changed during notable time periods, such as the beginning of the Covid-19 Pandemic, after September 11th 2001, or the 2007-2008 financial crisis.
- Comparing the rate during the presidencies of George W. Bush, Barack Obama, and Donald Trump.

We will focus on the last one.

```python
bush = euro_to_dollar[
    (euro_to_dollar['Time'] > '2001-01-20') & 
    (euro_to_dollar['Time'] < '2009-01-20')
]
obama = euro_to_dollar[
    (euro_to_dollar['Time'] > '2009-01-20') & 
    (euro_to_dollar['Time'] < '2017-01-20')
    
]
trump = euro_to_dollar[
    (euro_to_dollar['Time'] > '2017-01-20') & 
    (euro_to_dollar['Time'] < '2021-01-20')
]

presidencies = [bush,obama,trump]
labels = ["Bush","Obama"," Trump"]
colors = ["purple","orange","blue"]


style.use("fivethirtyeight")
fig, ax = plt.subplots(1, 3,figsize = (40,10),tight_layout = True)

### Adding a title and a subtitle
fig.text(0, 1.15, "EURO-USD rate increased under Bush and Trump, decreased under Obama",
         fontsize=50, weight='bold')
fig.text(0, 1.05, '''EURO-USD exchange rates under George W. Bush (2001 - 2009), Barack Obama (2009-2017),
and Donald Trump (2017-2021)''',
        fontsize=30)
# fig.text(
#         x=,y = 1.4,
#         s="EURO-USD Rate Increased under Bush, Trump, Decreased Under Obama",
#         size = 26,
#         weight = 'bold'
#     )
# fig.text(
#         x =0, y = 1.3,
#         s = "EURO-USD exchange rates during last 3 presidencies (2001-2021)),",
#         size = 22,
#         color = 'gray'
#     )
# fig.text(
#         x =0, y = 1.2,
#         s = "and Donald Trump (2017-2021)",
#         size = 22,
#         color = 'gray'
#     )
# plt.figure(figsize = (20,3))

for i in range(3):
#     plt.subplot(1,3,i+1)
    ax[i].plot(presidencies[i]['Time'],presidencies[i]['Rolling_mean'],label = labels[i],color = colors[i])
#     ax[i].xticks(rotation = 30)
    ax[i].set_title(labels[i],color = colors[i],fontsize = 40)
    ax[i].set_ylim(.8,1.6)

# ax[0].text(731516.0,1.65,size = 30, s = "George W. Bush",color = colors[0])
# ax[0].text(731750.0,1.61,size = 20, s = "(2001-2009)",color = 'gray')

# ax[1].text(734288.0,1.65,size = 30, s = "Barack Obama",color = colors[1])
# ax[1].text(734500.0,1.61,size = 20, s = "(2009-2017)",color = 'gray')

# ax[2].text(736855.0,1.65,size = 30, s = "Donald Trump",color = colors[2])
# ax[2].text(736950.0,1.61,size = 20, s = "(2017-2021)",color = 'gray')


    
plt.show()
# plt.figure(figsize = (20,3))

# bush_thru_trump = euro_to_dollar[
#     euro_to_dollar["Time"].between('2001-01-20', '2021-01-20')
# ]
# plt.plot(bush_thru_trump['Time'],bush_thru_trump['Rolling_mean'])

plt.show()
```

We see that the Euro overall gained strenth during Bush's presidency by a net of .76.  It then loast about 60% of that (0.48) during Obama's term, and stayed comparitively flat but gaining back nearly .2 points during Trump's four years in office.  It is interesting to note that the Euro generally got stronger during the two Republicans' terms while losing strength during the one Democrat's.  Without more data it's difficult to say whether this is a coincidence or if the party in power contributes to the relative strength of the euro and the dollar.  According to these articles on  [fxempire](https://www.fxempire.com/news/article/the-effect-of-us-presidents-on-eurusd-movement-393715) and [Walton College](https://walton.uark.edu/insights/the-almighty-dollar.php), it appears these trends have existed during the existence of the euro, largely due to the tighter monetary policies of Republican administrations and the looser policies of the Democtrats.

```python
      
euro_to_dollar['President'] = np.nan       

euro_to_dollar.loc[
    (euro_to_dollar['Time'] >= '2001-01-20') & 
    (euro_to_dollar['Time'] < '2009-01-20')
    ,"President"] = "Bush"

euro_to_dollar.loc[
    (euro_to_dollar['Time'] >= '2009-01-20') & 
    (euro_to_dollar['Time'] < '2017-01-20')
    ,"President"] = "Obama"

euro_to_dollar.loc[
    (euro_to_dollar['Time'] > '2017-01-20') & 
    (euro_to_dollar['Time'] < '2021-01-20')
    ,"President"] = "Trump"

euro_to_dollar["President"].value_counts()

```
